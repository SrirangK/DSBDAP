```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/ducati-corse-dtwgc0000301-analog-watch-men/product-reviews/itm29f88aa7861be?pid=WATGJ769XJNDMBKK&lid=LSTWATGJ769XJNDMBKKI4ASMH&marketplace=FLIPKART')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.find_all('div', {'class': '_11pzQk'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
good_ratings = soup.find_all('div', {'class': 'XQDdHH Ga3i8K _9lBNRY'})
for rating in good_ratings:
    print(rating.get_text() + "  ")

bad_ratings = soup.find_all('div', {'class': 'XQDdHH Czs3gR Ga3i8K _9lBNRY'})
for rating in bad_ratings:
    print(rating.get_text() + "  ")
```

```python
overall_rating = soup.find('div', {'class': 'XQDdHH E3wXvA'}).get_text()
print(overall_rating)
```

```python
tags = soup.find_all('p', {'class': 'MztJPv'})
for tag in tags:
    print(tag.get_text() + "  ")
```

```python
customers = soup.find_all('p', {'class': '_2NsDsF AwS1CA MDcJkH'})
for customer in customers:
    print(customer.get_text() + "  ")
```
