```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DSBDALExam DataSets/Adult/adult_dataset.csv')
df
```

```python
plt.hist(df['age'],bins=20,edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
edu_counts = df['education'].value_counts()
plt.plot(edu_counts.values,edu_counts.index,  'o', color='purple')
plt.title('Dot Plot of Education Levels')
plt.xlabel('Count')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='income', palette='pastel')
plt.title('Income Group Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
avg_hours = df.groupby('age')['hours-per-week'].mean()

plt.figure(figsize=(10, 5))
plt.plot(avg_hours.index, avg_hours.values, color='green')
plt.title('Average Hours per Week by Age')
plt.xlabel('Age')
plt.ylabel('Avg Hours per Week')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['hours-per-week'], color='black', width=0.4)
sns.histplot(df['hours-per-week'], bins=20, color='lightgreen', alpha=0.5, kde=False)
sns.stripplot(x=df['hours-per-week'], color='lightblue', size=4, jitter=0.2)
plt.title('Box Plot with Histogram and Scatter for Hours-per-week')
plt.xlabel('Hours per Week')
plt.grid(True)
plt.show()
```
