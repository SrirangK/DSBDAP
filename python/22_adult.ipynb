```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('DSBDALExam DataSets/Adult/adult_dataset.csv')
df
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['age'], bins=30, color='skyblue')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.histplot(df['hours-per-week'], bins=30, color='salmon')
plt.title('Histogram of Hours-per-week')
plt.xlabel('Hours')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
income_counts = df['income'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(income_counts.values, labels=income_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Income Distribution')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='income', y='hours-per-week', data=df, palette='Set2')
plt.title('Box Plot of Hours per Week by Income Group')
plt.xlabel('Income')
plt.ylabel('Hours per Week')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='age', y='hours-per-week', hue='income', alpha=0.6)
plt.title('Age vs Hours per Week by Income')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.grid(True)
plt.show()
```

```python
sns.set(style="whitegrid")
g = sns.jointplot(data=df, x="age", y="hours-per-week", hue="income", kind="scatter", height=7, palette='Set1')

sns.boxplot(data=df, x="age", ax=g.ax_marg_x, color='lightblue')
sns.boxplot(data=df, y="hours-per-week", ax=g.ax_marg_y, color='lightgreen',orient='v')

plt.suptitle('Scatter Plot with Marginal Boxplots: Age vs Hours-per-week', y=1.02)
plt.show()
```
