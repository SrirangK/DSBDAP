```python
import pandas as pd
```

```python
df=pd.read_csv('DSBDALExam DataSets/forestfires/forestfires.csv')
df
```

```python
not_affected = df[df['area'] == 0]
partially_affected = df[(df['area'] > 0) & (df['area'] <= 10)]
mostly_affected = df[df['area'] > 10]
```

```python
merged_subset = pd.merge(partially_affected, mostly_affected,on=['X','Y','month','day'])
merged_subset
```

```python
sorted_df = df.sort_values(by=['temp', 'wind', 'area'], ascending=[True, True, False])
sorted_df
```

```python
transposed_df = df.transpose()
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['X','Y','month', 'day'], value_vars=['temp', 'wind', 'area'],
                    var_name='Measurement', value_name='Value')
melted_df
```

```python
casted_df = melted_df.pivot_table(index=['X','Y','month', 'day'], columns='Measurement', values='Value', aggfunc='mean').reset_index()
casted_df
```
