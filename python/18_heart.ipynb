```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DSBDALExam DataSets/HeartDisease/heart.csv')
df.head()
```

```python
df.hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms: Distribution of Numerical Features", fontsize=16)
# plt.tight_layout()
plt.show()
```

```python
sizes = df['target'].value_counts()
sizes
```

```python
labels = ['Disease','No Disease']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Heart Disease Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['age', 'chol']], palette='Set2')
plt.title('Box Plots: Age and Cholesterol')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='target', palette='Set1')
plt.title('Scatter Plot: Age vs Cholesterol by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```

```python
g = sns.jointplot(data=df, x='age', y='chol', hue='target', kind='scatter')

sns.boxplot(data=df, x='age', ax=g.ax_marg_x, color='lightblue')
sns.boxplot(data=df, y='chol', ax=g.ax_marg_y, color='lightgreen', orient='v')
```
