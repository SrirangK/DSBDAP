```python
melted_df = pd.melt(
    df,
    id_vars=['Type'],
    value_vars=['comment', 'like', 'share'],
    var_name='Interaction Type',
    value_name='Count'
)
melted_df
```

```python
cast_df = melted_df.pivot_table(
    index=['Type', 'Post Month'],
    columns='Interaction Type',
    values='Count',
    aggfunc='sum'
).reset_index()
cast_df
```
