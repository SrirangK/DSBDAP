```python
df_melted = pd.melt(df,
                    id_vars=['title'], 
                    value_vars=['budget', 'revenue', 'runtime'],
                    var_name='feature',
                    value_name='value')
```

```python
df_melted
```

```python
df_cast=df_melted.pivot_table(
    index=['title'],
    columns='feature',
    values='value',
    aggfunc='first'
    )

df_cast.sort_values(by='revenue',ascending=False).transpose()
# df_cast.transpose()
```
