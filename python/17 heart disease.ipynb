```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DSBDALExam DataSets/HeartDisease/heart.csv')
df
```

```python
df_sort=df.sort_values(by='trestbps',ascending=False)
df_sort
```

```python
df.replace('?', np.nan, inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)
```

```python
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)
```

```python
plt.figure(figsize=(12, 10))
numcols=['age','trestbps','chol','thalach']
df[numcols].hist( bins=20, color='skyblue', edgecolor='black')
# plt.title('Histogram of Age')
# plt.xlabel('Age')
# plt.ylabel('Frequency')
# plt.grid(True)
plt.show()
```

```python
plt.hist(df['age'],bins=20,edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['trestbps'],df['thalach'], 'o', color='red')
plt.title('Dot Plot of Resting Blood Pressure (trestbps)')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('thalach')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# sns.countplot(x='cp', data=df)
cpc=df['cp'].value_counts()
plt.bar(x=cpc.index,height=cpc.values)
plt.xticks(ticks=cpc.index)
plt.title('Bar Plot of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
df_sorted = df.sort_values('age')
plt.figure(figsize=(8, 5))
plt.plot(df_sorted['age'], df_sorted['thalach'], color='green')
plt.title('Line Chart of Max Heart Rate (thalach) vs Age')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (thalach)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# sns.boxplot(x=df['age'], color='lightblue', width=0.5)
df['age'].hist(bins=20, color='lightgreen', alpha=0.5)
sns.boxplot(x=df['age'], color='lightblue', width=0.5)
sns.stripplot(data=df,x='age', color='red')
plt.title('Box Plot with Histogram and Scatter for Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```
