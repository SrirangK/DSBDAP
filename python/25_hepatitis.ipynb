```python
import pandas as pd
```

```python
df=pd.read_csv("DSBDALExam DataSets/Hepatitis/hepatitis_csv.csv")
df
```

```python
male_subset = df[df['sex'] == 'male']
female_subset = df[df['sex'] == 'female']
```

```python
merged_df = pd.merge(male_subset, female_subset,on='age')
merged_df
```

```python
sorted_df = df.sort_values(by=['age', 'sgot', 'protime'])
sorted_df
```

```python
transposed_df = sorted_df.transpose()
```

```python
melted_df = pd.melt(df, id_vars=['sex'], value_vars=['age', 'sgot', 'protime'])
melted_df
```

```python
wide_df = melted_df.pivot_table(index='sex', columns='variable', values='value', aggfunc='mean')
wide_df
```
